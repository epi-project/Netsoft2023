```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker 
```

## Obtain history data

```python
df = pd.read_csv('k8s_historical_states_discrete.csv')
results = df[['timestep', 'reward', 'number_of_pods', 'latency', 'info', 'hpa_threshold']]
```

## Plot the tendency in training period with -u=50

```python
podsTrainVis = results[1005:][['timestep', 'number_of_pods', 'info']]
latencyTrainVis = results[1005:][['timestep', 'latency', 'info']]
thesTrainVis = results[1005:][['timestep', 'hpa_threshold', 'info']]
rewardTrainVis = results[1005:][['timestep', 'reward', 'info']]
```

```python
plt.figure(figsize=(15, 12))
plt.subplot(2, 2, 1)
# plt.figure(dpi=120)
plt.xlabel('round')
plt.ylabel('number_of_pods')
plt.title('changes in number of pods during training')
plt.plot(podsTrainVis['timestep'], podsTrainVis['number_of_pods'])
plt.vlines(35, 0, max(podsTrainVis['number_of_pods']), colors = "r", linestyles = "dashed")
plt.subplot(2, 2, 2)
# plt.figure(dpi=120)
plt.xlabel('round')
plt.ylabel('hpa_threshold')
plt.title('changes in hpa_threshold during training')
# 
def to_percent(temp, position):
    return '%1.0f'%(temp) + '%'
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(to_percent))
plt.plot(thesTrainVis['timestep'], thesTrainVis['hpa_threshold'])
plt.vlines(35, 0, max(thesTrainVis['hpa_threshold']), colors = "r", linestyles = "dashed")
# plt.show()
# plt.figure(figsize=(15, 5))
plt.subplot(2, 2, 3)
plt.xlabel('round')
plt.ylabel('reward')
plt.title('changes in rewards during training')
plt.plot(rewardTrainVis['timestep'], rewardTrainVis['reward'])
plt.vlines(35, 0, max(rewardTrainVis['reward']), colors = "r", linestyles = "dashed")
plt.subplot(2, 2, 4)
plt.xlabel('round')
plt.ylabel('latency/seconds')
plt.title('changes in latency during training')
plt.gca().yaxis.set_major_formatter(mticker.FormatStrFormatter('%.1f s'))
plt.plot(latencyTrainVis['timestep'], latencyTrainVis['latency'])
plt.vlines(35, 0, max(latencyTrainVis['latency']), colors = "r", linestyles = "dashed")
```

```python
plt.figure(figsize=(15, 12))
plt.subplot(2, 2, 1)
# plt.figure(dpi=120)
plt.xlabel('round')
plt.ylabel('number_of_pods')
plt.title('changes in number of pods during testing')
plt.vlines(48, 0, max(podsTestVis['number_of_pods']), colors = "r", linestyles = "dashed")
plt.vlines(66, 0, max(podsTestVis['number_of_pods']), colors = "r", linestyles = "dashed")
plt.vlines(91, 0, max(podsTestVis['number_of_pods']), colors = "r", linestyles = "dashed")
plt.vlines(120, 0, max(podsTestVis['number_of_pods']), colors = "r", linestyles = "dashed")
plt.plot(podsTestVis['timestep'], podsTestVis['number_of_pods'])
plt.subplot(2, 2, 2)
# plt.figure(dpi=120)
plt.xlabel('round')
plt.ylabel('hpa_threshold')
plt.title('changes in hpa_threshold during testing')
# 
def to_percent(temp, position):
    return '%1.0f'%(temp) + '%'
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(to_percent))
plt.vlines(48, 0, max(thesTestVis['hpa_threshold']), colors = "r", linestyles = "dashed")
plt.vlines(66, 0, max(thesTestVis['hpa_threshold']), colors = "r", linestyles = "dashed")
plt.vlines(91, 0, max(thesTestVis['hpa_threshold']), colors = "r", linestyles = "dashed")
plt.vlines(120, 0, max(thesTestVis['hpa_threshold']), colors = "r", linestyles = "dashed")
plt.plot(thesTestVis['timestep'], thesTestVis['hpa_threshold'])
# plt.show()
# plt.figure(figsize=(15, 5))
plt.subplot(2, 2, 3)
plt.xlabel('round')
plt.ylabel('reward')
plt.title('changes in rewards during testing')
plt.vlines(48, 0, max(rewardTestVis['reward']), colors = "r", linestyles = "dashed")
plt.vlines(66, 0, max(rewardTestVis['reward']), colors = "r", linestyles = "dashed")
plt.vlines(91, 0, max(rewardTestVis['reward']), colors = "r", linestyles = "dashed")
plt.vlines(120, 0, max(rewardTestVis['reward']), colors = "r", linestyles = "dashed")
plt.plot(rewardTestVis['timestep'], rewardTestVis['reward'])
plt.subplot(2, 2, 4)
plt.xlabel('round')
plt.ylabel('latency/seconds')
plt.title('changes in latency during testing')
plt.vlines(48, 0, max(latencyTestVis['latency']), colors = "r", linestyles = "dashed")
plt.vlines(66, 0, max(latencyTestVis['latency']), colors = "r", linestyles = "dashed")
plt.vlines(91, 0, max(latencyTestVis['latency']), colors = "r", linestyles = "dashed")
plt.vlines(120, 0, max(latencyTestVis['latency']), colors = "r", linestyles = "dashed")
plt.gca().yaxis.set_major_formatter(mticker.FormatStrFormatter('%.1f s'))
plt.plot(latencyTestVis['timestep'], latencyTestVis['latency'])
```
